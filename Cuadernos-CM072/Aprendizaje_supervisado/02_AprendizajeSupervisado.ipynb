## Aprendizaje supervisado

El aprendizaje supervisado se usa cuando queremos predecir un resultado determinado a partir de una entrada determinada, y tenemos ejemplos de pares de entrada/salida. 

Construimos un modelo de Machine Learning a partir de estos pares de entrada/salida, que comprenden nuestro conjunto de entrenamiento. 

Nuestro objetivo es hacer predicciones precisas para datos nuevos nunca antes vistos. El aprendizaje supervisado a menudo requiere esfuerzo humano para construir el conjunto de entrenamiento, pero luego automatiza y a menudo acelera una tarea que de otra manera sería laboriosa o inviable.

### Clasificación y Regresión

En la clasificación, el objetivo es predecir una etiqueta de clase, que es una elección de una lista predefinida de posibilidades.

La clasificación a veces se separa en clasificación binaria, que es el caso especial de distinguir exactamente entre dos clases, y la clasificación multiclase, que es la clasificación entre más de dos clases. Puede pensar en la clasificación binaria como tratar de responder una pregunta sí/no.

Clasificar correos electrónicos como correo no deseado o no,  es un ejemplo de un problema de clasificación binaria. En esta tarea de clasificación binaria, la pregunta sí/no que se haría sería  es: ¿Es este correo electrónico no deseado?.

En la clasificación binaria, a menudo hablamos de una clase que es la clase positiva y la otra clase es la clase negativa. Aquí, lo positivo no representa tener beneficio o valor, sino más bien cuál es el objeto del estudio. Entonces, cuando se busca spam, "positivo" podría significar la clase de spam. Cuál de las dos clases se llama positiva es a menudo una cuestión subjetiva y específica del dominio.

El ejemplo del iris, por otro lado, es un ejemplo de un problema de clasificación multiclase. Otro ejemplo es predecir en qué idioma se encuentra un sitio web a partir del texto del sitio web. Las clases aquí serían una lista predefinida de idiomas posibles.

Para las tareas de regresión, el objetivo es predecir un número continuo, o un número de coma flotante en términos de programación (o número real en términos matemáticos). La predicción del ingreso anual de una persona a partir de su educación, su edad y el lugar donde vive es un ejemplo de una tarea de regresión. Al predecir el ingreso, el valor predicho es una cantidad, y puede ser cualquier número en un rango dado.

Otro ejemplo de una tarea de regresión es predecir el rendimiento de una granja de maíz dados los atributos como los rendimientos anteriores, el clima y la cantidad de empleados que trabajan en la granja. El rendimiento nuevamente puede ser un número arbitrario.


Una manera fácil de distinguir entre tareas de clasificación y regresión es preguntar si hay algún tipo de continuidad en el resultado. Si hay continuidad entre posibles resultados, entonces el problema es un problema de regresión. Piensa en predecir el ingreso anual.

Hay una continuidad clara en la salida. Si una persona gana `S/.40,000` o `S./40,001` al año no hace una diferencia tangible, a pesar de que se trata de cantidades de dinero diferentes; si nuestro algoritmo predice `S/.39,999` o `S/.40.001` cuando debería haber predicho `S/.40,000`, no nos importa mucho.

Por el contrario, para la tarea de reconocer el idioma de un sitio web (que es un problema de clasificación), no hay materia de grado. Un sitio web está en un idioma o en otro. No hay continuidad entre idiomas, y no hay ningún idioma entre inglés y francés.

### Generalización, sobrefijado y subfijado

En el machine learning, queremos construir un modelo sobre los datos de entrenamiento y luego poder hacer predicciones precisas sobre datos nuevos y no vistos  que tengan las mismas características que el conjunto de entrenamiento que utilizamos. 

Si un modelo puede hacer predicciones precisas sobre datos no vistos, decimos que es capaz de generalizar desde el conjunto de entrenamiento hasta el conjunto de prueba. Queremos construir un modelo que sea capaz de generalizar con la mayor precisión posible.

Usualmente construimos un modelo de tal manera que pueda hacer predicciones precisas en el conjunto de entrenamiento. Si los conjuntos de entrenamiento y prueba tienen suficiente en común, esperamos que el modelo también sea preciso en el conjunto de prueba. Sin embargo, hay algunos casos en que esto puede salir mal.

Por ejemplo, si nos permitimos construir modelos muy complejos, siempre podemos ser tan precisos como nos gusta en el conjunto de entrenamiento.

#### Un ejemplo

Digamos que un científico de datos novato quiere predecir si un cliente comprará un bote, dado los registros de compradores de embarcaciones anteriores y clientes que sabemos que no están interesados en comprar un bote. El objetivo es enviar correos electrónicos promocionales a las personas que probablemente realicen una compra, pero no molestar a los clientes que no estarán interesados.

En el mundo real, este es un problema complicado. Si bien sabemos que los otros clientes todavía no nos han comprado un bote, es posible que hayan comprado uno de otra persona, o que aún estén ahorrando y planeen comprar uno en el futuro.

![](Grafico1.png)


Queremos saber si es probable que nuevos clientes compren un bote. Por lo tanto, deseamos encontrar una regla que funcione bien para los nuevos clientes, y lograr una precisión del 100 por ciento en el conjunto de entrenamiento no nos ayuda allí. 

Es posible que no esperemos que la regla que nos dio nuestro científico de datos funcione muy bien para los nuevos clientes. Parece demasiado complejo y está respaldado por muy pocos datos. Por ejemplo, la parte "está divorciada" de la regla depende de un solo cliente.

La única medida de si un algoritmo tendrá un buen rendimiento en los datos nuevos es la evaluación en el conjunto de prueba. Sin embargo, intuitivamente esperamos que los modelos simples se generalicen mejor a los datos nuevos. Si la regla era "Las personas mayores de 50 años quieren comprar un bote", y esto explicaría el comportamiento de todos los clientes, confiaríamos en él más que en la regla que involucra a los niños, el estado civil además de la edad.


Construir un modelo que es demasiado complejo para la cantidad de información que tenemos se llama **sobreajuste**.

El **sobreajuste** se produce cuando se ajusta un modelo demasiado a las particularidades del conjunto de entrenamiento y se obtiene un modelo que funciona bien en el conjunto de entrenamiento, pero que no puede generalizar a datos nuevos.


Por otro lado, si el modelo es demasiado simple, por ejemplo, "todos los propietarios de una casa compran un bote", es posible que no pueda capturar todos los aspectos y la variabilidad de los datos, y su modelo funcionará mal incluso en el conjunto de entrenamiento. Elegir un modelo demasiado simple se llama **subfijado**.


Cuanto más complejos permitamos que nuestro modelo sea, mejor podremos predecir en los datos de entrenamiento. Sin embargo, si nuestro modelo se vuelve demasiado complejo, comenzamos a centrarnos demasiado en cada punto de datos individual en nuestro conjunto de entrenamiento, y el modelo no se generalizará bien a los datos nuevos.

Hay un punto intermedio entre ellos que producirá el mejor rendimiento de generalización. Este es el modelo que queremos encontrar. En la siguiente figura se ilustra la relación entre el sobreajuste y el subajuste.

### Relación de la complejidad del modelo con el tamaño del conjunto de datos

Es importante tener en cuenta que la complejidad del modelo está íntimamente ligada a la variación de las entradas contenidas en el conjunto de datos de entrenamiento: cuanto mayor sea la variedad de puntos de datos que contenga el conjunto de datos, más complejo será el modelo que pueda usar sin sobreajustar. 

Por lo general, la recopilación de más puntos de datos producirá más variedad, por lo que los conjuntos de datos más grandes permiten construir modelos más complejos. Sin embargo, simplemente duplicar los mismos puntos de datos o recopilar datos muy similares no ayudará.

Volviendo al ejemplo de venta de embarcaciones, si vemos 10.000 filas más de datos de clientes, y todos cumplieron con la regla "Si el cliente es mayor de 45 años y tiene menos de 3 hijos o no está divorciado, entonces ellos quieren compre un bote", sería mucho más probable que creamos que esta es una buena regla que cuando se desarrolló usando solo las 12 filas como se muestrá en una tabla anterior.

Utilizaremos varios conjuntos de datos para ilustrar los diferentes algoritmos. Algunos de los conjuntos de datos serán pequeños y sintéticos (es decir, inventados), diseñados para resaltar aspectos particulares de los algoritmos. Otros conjuntos de datos serán grandes, ejemplos del mundo real.

Un ejemplo de un conjunto de datos de clasificación de dos clases sintéticos es el conjunto de datos `front`, que tiene dos características. El siguiente código crea un diagrama de dispersión que visualiza todos los puntos de datos en este conjunto de datos.

```python
%matplotlib inline
import numpy as np
from sklearn.datasets.samples_generator import make_blobs

def hacer_forge():
    # un conjunto de datos diseñado a mano
    X, y = make_blobs(centers=2, random_state=4, n_samples=30)
    y[np.array([7, 27])] = 0
    mascara = np.ones(len(X), dtype=np.bool)
    mascara[np.array([0, 1, 5, 26])] = 0
    X, y = X[mascara], y[mascara]
    return X, y
```

Para ilustrar los algoritmos de regresión, usaremos el conjunto de datos `wave`. El conjunto de datos `wave` tiene una única función de entrada y una variable de destino continua (o respuesta) que queremos modelar. El gráfico creado aquí muestra la característica única en el eje x y el objetivo de la regresión (el resultado) en el eje y:

```python
def hacer_wave(n_muestras=100):
    rnd = np.random.RandomState(42)
    x = rnd.uniform(-3, 3, size=n_muestras)
    y_no_ruido = (np.sin(4 * x) + x)
    y = (y_no_ruido + rnd.normal(size=len(x))) / 2
    return x.reshape(-1, 1), y
```

```python
X, y = hacer_wave(n_muestras=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel("Caracteristica")
plt.ylabel("Objetivo")
```

Estamos utilizando estos conjuntos de datos muy simples y de baja dimensión porque los podemos visualizar fácilmente: una página impresa tiene dos dimensiones, por lo que los datos con más de dos características son difíciles de mostrar. Cualquier intuición derivada de conjuntos de datos con algunas características (también llamadas conjuntos de datos de baja dimensión) podría no mantenerse en conjuntos de datos con muchas características (conjuntos de datos de alta resolución).

## K-vecinos más cercanos

El algoritmo k-NN es posiblemente el algoritmo del machine learning más simple. La construcción del modelo consiste únicamente en almacenar el conjunto de datos de entrenamienti Para hacer una predicción para un nuevo punto de datos, el algoritmo encuentra los puntos de datos más cercanos en los conjuntos de datos de entrenamiento, sus "vecinos más cercanos".

### Clasificación k-vecinos

En su versión más simple, el algoritmo k-NN solo considera exactamente un vecino más cercano, que es el punto de datos de entrenamiento más cercano al punto que queremos hacer una predicción.

La predicción es simplemente la salida conocida para este punto de entrenamiento. Se ilustra esto para el caso de la clasificación en el conjunto de datos `forge`:


```python
from sklearn.metrics import euclidean_distances
from sklearn.neighbors import KNeighborsClassifier

def dibuja_knn_clasificacion(n_vecinos=1):
    X, y = hacer_forge()

    X_prueba = np.array([[8.2, 3.66214339], [9.9, 3.2], [11.2, .5]])
    distancia = euclidean_distances(X, X_prueba)
    cercania = np.argsort(distancia, axis=0)

    for x, vecinos in zip(X_prueba, cercania.T):
        for vecino in vecinos[:n_vecinos]:
            plt.arrow(x[0], x[1], X[vecino, 0] - x[0],
                      X[vecino, 1] - x[1], head_width=0, fc='k', ec='k')

    clasificador_knn = KNeighborsClassifier(n_neighbors=n_vecinos).fit(X, y)
    puntos_prueba = dibuja_dispersion_discreta(X_prueba[:, 0], X_prueba[:, 1], 
                                        clasificador_knn.predict(X_prueba),marcadores="*")
    puntos_entrenamientos = dibuja_dispersion_discreta(X[:, 0], X[:, 1], y)
    plt.legend(puntos_entrenamientos + puntos_prueba, ["Clase de entrenamiento 0", "Clase de entrenamiento 1", 
                                                       "pred prueba 0", "pred prueba 1"])
```

```python
dibuja_knn_clasificacion(n_vecinos=1)
```

Aquí, agregamos tres nuevos puntos de datos, que se muestran como estrellas. Para cada uno de ellos, marcamos el punto más cercano en el conjunto de entrenamiento. La predicción del algoritmo de un vecino más cercano es la etiqueta de ese punto. 

En lugar de considerar solo al vecino más cercano, también podemos considerar un número arbitrario, k, de vecinos. Aquí es donde proviene el nombre del algoritmo k-vecinos más cercano.

Cuando consideramos a más de un vecino, usamos el voto para asignar una etiqueta. Esto significa que para cada punto de prueba, contamos cuántos vecinos pertenecen a la clase 0 y cuántos vecinos pertenecen a la clase 1. Luego asignamos la clase que es más frecuente: en otras palabras, la clase mayoritaria entre los k vecinos más cercanos. El siguiente ejemplo se usa los tres vecinos más cercanos:

```python
dibuja_knn_clasificacion(n_vecinos=3)
```

Si bien esta ilustración es para un problema de clasificación binaria, este método se puede aplicar a conjuntos de datos con cualquier cantidad de clases. Para más clases, contamos cuántos vecinos pertenecen a cada clase y de nuevo predecimos la clase más común.

#### El algoritmo k-vecinos más cercanos y scikit learn

En primer lugar, dividimos nuestros datos en un conjunto de entrenamiento y prueba para que podamos evaluar el rendimiento de generalización.


```python
from sklearn.model_selection import train_test_split
X, y = hacer_forge()
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, random_state=0)
```

A continuación, importamos e instanciamos la clase. Aquí es cuando podemos establecer parámetros, como la cantidad de vecinos a usar. Aquí, lo configuramos en 3:

```python
from sklearn.neighbors import KNeighborsClassifier
clasificador_knn = KNeighborsClassifier(n_neighbors=3)
```

Ahora, ajustamos el clasificador usando el conjunto de entrenamiento. Para `KNeighborsClassifier` esto significa almacenar el conjunto de datos, por lo que podemos calcular los vecinos durante la predicción:

```python
clasificador_knn.fit(X_entrenamiento, y_entrenamiento)
```

Para hacer predicciones sobre los datos de prueba, llamamos al método `predict`.

Para cada punto de datos en el conjunto de prueba, este calcula sus vecinos más cercanos en el conjunto de entrenamiento y encuentra la clase más común entre estos:

```python
print("Predicciones del conjunto de pruebas: {}".format(clasificador_knn.predict(X_prueba)))
```

Para evaluar qué tan bien generaliza nuestro modelo, podemos llamar al método `score` con los datos de prueba junto con las etiquetas de prueba:

```python
print("Precision del conjunto de pruebas: {:.2f}".format(clasificador_knn.score(X_prueba, y_prueba)))
```

#### Otro ejemplo

Utilizaremos el conjunto de datos `cancer` del mundo real. Comenzamos dividiendo el conjunto de datos en un conjunto de entrenamiento y un conjunto de prueba. Luego evaluamos el entrenamiento y el rendimiento del conjunto de prueba con diferentes números de vecinos.

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split( cancer.data, cancer.target, 
                                                    stratify=cancer.target, random_state=66)
precision_entrenamiento = []
precision_prueba = []

# Probamos  n_vecinos desde 1 al 10
conf_vecinos = range(1, 11)

# Construimos el modelo
for n_vecinos in conf_vecinos:
    clasificador_knn = KNeighborsClassifier(n_neighbors=n_vecinos)
    clasificador_knn.fit(X_entrenamiento, y_entrenamiento)

    # registrando la precision del conjunto de entrenamiento
    precision_entrenamiento.append(clasificador_knn.score(X_entrenamiento, y_entrenamiento))
    # registrando la precision del conjunto de pruebas
    precision_prueba.append(clasificador_knn.score(X_prueba, y_prueba))

plt.plot(conf_vecinos, precision_entrenamiento, label="precision de entrenamiento")
plt.plot(conf_vecinos, precision_prueba, label="precision de prueba")

plt.ylabel("Precision")
plt.xlabel("n_vecinos")
plt.legend()
```

El gráfico muestra la precisión del entrenamiento y del conjunto de pruebas en el eje y frente al ajuste de n_neighbors en el eje x. Considerando un vecino más cercano, la predicción en el conjunto de entrenamiento es perfecta. Pero cuando se consideran más vecinos, el modelo se simplifica y la precisión del entrenamiento disminuye.

La precisión del conjunto de pruebas para usar un solo vecino es menor que cuando se usan más vecinos, lo que indica que usar el vecino más cercano conduce a un modelo que es demasiado complejo. Por otro lado, cuando se consideran 10 vecinos, el modelo es demasiado simple y el rendimiento es aún peor. El mejor rendimiento está en el medio, usando alrededor de seis vecinos.

Aún así, es bueno tener en cuenta la escala del gráfico. El peor rendimiento es aproximadamente el 88% de precisión, que aún podría ser aceptable.

### Regresión de k-vecinos

También hay una variante de regresión del algoritmo k-vecino más cercano. De nuevo, comencemos utilizando el vecino más cercano, esta vez usando el conjunto de datos `wave`. Agregamos tres puntos de datos de prueba como estrellas verdes en el eje x. La predicción usando un solo vecino es solo el valor objetivo del vecino más cercano. Estos se muestran como estrellas azules.

```python
from sklearn.neighbors import KNeighborsRegressor
from matplotlib.colors import ListedColormap
cm3 = ListedColormap(['#0000aa', '#ff2020', '#50ff50'])

def dibuja_regresion_knn(n_vecinos=1):
    X, y = hacer_wave(n_muestras=40)
    X_prueba = np.array([[-1.5], [0.9], [1.5]])

    distancia = euclidean_distances(X, X_prueba)
    cercania = np.argsort(distancia, axis=0)

    plt.figure(figsize=(10, 6))

    reg_knn = KNeighborsRegressor(n_neighbors=n_vecinos).fit(X, y)
    y_pred = reg_knn.predict(X_prueba)

    for x, y_, vecinos in zip(X_prueba, y_pred, cercania.T):
        for vecino in vecinos[:n_vecinos]:
                plt.arrow(x[0], y_, X[vecino, 0] - x[0], y[vecino] - y_, head_width=0, fc='k', ec='k')

    entrenamiento, = plt.plot(X, y, 'o', c=cm3(0))
    prueba, = plt.plot(X_prueba, -3 * np.ones(len(X_prueba)), '*', c=cm3(2),
                     markersize=20)
    pred, = plt.plot(X_prueba, y_pred, '*', c=cm3(0), markersize=20)
    plt.vlines(X_prueba, -3.1, 3.1, linestyle="--")
    plt.legend([entrenamiento, prueba, pred],
               ["Entrenamiento data/objetivo", "dato prueba", "prediccion prueba"],
               ncol=3, loc=(.1, 1.025))
    plt.ylim(-3.1, 3.1)
    plt.xlabel("Caracteristica")
    plt.ylabel("Objetivo")
```

```python
dibuja_regresion_knn(n_vecinos=1)
```

Nuevamente, podemos usar más que un único vecino más cercano para la regresión. Al usar múltiples vecinos más cercanos, la predicción es el promedio, o la media, de los vecinos relevantes.

```python
dibuja_regresion_knn(n_vecinos=3)
```

El algoritmo de k-vecinos vecinos para la regresión se implementa en la clase `KNeighborsRegressor` en scikit-learn. Se usa de manera similar a KNeighborsClassifier:

```python
from sklearn.neighbors import KNeighborsRegressor
X, y = hacer_wave(n_muestras=40)

# dividimos el conjunto de datos wave en un conjunto de entrenamiento y una prueba 
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X, y, random_state=0)

# instanciamos  el model y consideramos el numero de vecinos a 3
reg_knn = KNeighborsRegressor(n_neighbors=3)

# fijamos el modelo usando el conjunto de entrenamiento y objetivos de entrenamiento 
reg_knn.fit(X_entrenamiento, y_entrenamiento)
```

Ahora hacemos predicciones sobre el conjunto de pruebas:

```python
print("Predicciones del conjunto de entrenamiento:\n{}".format(reg_knn.predict(X_prueba)))
```

También podemos evaluar el modelo usando el método `score`, que para los regresores devuelve el puntaje $R^2$. 

El puntaje $R^2$, también conocido como el coeficiente de determinación, es una medida de bondad de una predicción para un modelo de regresión, y arroja un puntaje entre 0 y 1. 

Un valor de 1 corresponde a una predicción perfecta, y un valor de 0 corresponde a un modelo constante que solo predice la media de las respuestas del conjunto de entrenamiento, `y_train`:

```python
print("Conjunto prueba R^2: {:.2f}".format(reg_knn.score(X_prueba, y_prueba)))
```

### Analizando el regresor KNN

Para nuestro conjunto de datos unidimensional, podemos ver cuáles son las predicciones para todos los valores de características posibles. Para hacer esto, creamos un conjunto de datos de prueba que consta de muchos puntos en la línea:

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# creamos 1,000 puntos de datos, espaciados entre -3 y 3
linea = np.linspace(-3, 3, 1000).reshape(-1, 1)

# hacemos predicciones utilizando 1, 3, o 9 vecinos
for n_vecinos, ax in zip([1, 3, 9], axes):
    reg_knn = KNeighborsRegressor(n_neighbors=n_vecinos)
    reg_knn.fit(X_entrenamiento, y_entrenamiento)
    ax.plot(linea, reg_knn.predict(linea))
    ax.plot(X_entrenamiento, y_entrenamiento, '^', c=cm2(0), markersize=8)
    ax.plot(X_prueba, y_prueba, 'v', c=cm2(1), markersize=8)
    ax.set_title( "{} vecinos\nresultado entrenamiento: {:.2f}\n resultado prueba: {:.2f}".format( n_vecinos,
                            reg_knn.score(X_entrenamiento, y_entrenamiento), reg_knn.score(X_prueba, y_prueba)))
    ax.set_xlabel("Caracteristica")
    ax.set_ylabel("Objetivo")
    axes[0].legend(["Prediccion de modelos ", "Entrenamiento datos\nobjetivo","Prueba datos\nobjetivo"], loc="best")
```

Como podemos ver en el gráfico, usando solo un vecino, cada punto en el conjunto de entrenamiento tiene una influencia obvia en las predicciones, y los valores predichos pasan por todos los puntos de datos. Esto lleva a una predicción muy inestable. Tener en cuenta a más vecinos lleva a predicciones más suaves, pero éstas tampoco se ajustan a los datos de entrenamiento.

### Fortalezas, debilidades y parámetros

En principio, hay dos parámetros importantes para el clasificador `KNeighbors`: el número de vecinos y cómo se mide la distancia entre los puntos de datos. En la práctica, usar una pequeña cantidad de vecinos como tres o cinco a menudo funciona bien, pero ciertamente se debe ajustar este parámetro. 

Elegir la medida correcta de distancia está algo más allá de cualquier alcancee. Por defecto, se usa la distancia euclidiana, que funciona bien en muchos entornos.

Una de las ventajas de k-NN es que el modelo es muy fácil de entender y a menudo, ofrece un rendimiento razonable sin muchos ajustes. El uso de este algoritmo es un buen método de referencia para probar, antes de considerar técnicas más avanzadas.

Construir el modelo de vecinos más cercanos suele ser muy rápido, pero cuando el conjunto de entrenamiento es muy grande (ya sea en número de características o en número de muestras), la predicción puede ser lenta. Al usar el algoritmo k-NN, es importante preprocesar sus datos.

Este enfoque a menudo no funciona bien en conjuntos de datos con muchas características (cientos o más), y lo hace particularmente mal con conjuntos de datos donde la mayoría de las características son 0 la mayoría de las veces (los denominados conjuntos de datos dispersos).

Entonces, aunque el algoritmo de k-vecinos más cercano es fácil de entender, no se usa a menudo en la práctica, debido a que la predicción es lenta y su incapacidad para manejar muchas características.
